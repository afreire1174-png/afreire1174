**1.3 Tipo de problema**

	Tipo escolhido: Classificação Binária

	Variável alvo:

	Evento_Crítico_Target
		SIM:
			Grau de risco Alto
			Grau de risco Crítico
		NÃO:
			Grau de risco Baixo
			Grau de risco Médio

O problema proposto neste MVP caracteriza-se como um problema de classificação supervisionada, pois a variável alvo da base de dados, Evento_Crítico_Target, é categórica e assume apenas duas classes possíveis: "Sim" e "Não".

A classe "Sim" representa ocorrências classificadas como eventos de Alto ou Crítico Grau de Risco, enquanto a classe "Não" representa ocorrências que não se enquadram nessa classificação. Assim, o objetivo do modelo é aprender, a partir das características históricas dos acidentes e quase-acidentes do setor elétrico, a identificar corretamente a qual dessas duas categorias uma nova ocorrência pertence.

Como a variável resposta não representa um valor numérico contínuo, mas sim categorias mutuamente exclusivas, o problema não é de regressão. Em vez de estimar uma quantidade, o modelo deverá decidir entre duas classes previamente definidas, caracterizando uma classificação binária supervisionada.

Além disso, a base de dados já possui registros históricos rotulados com a variável Evento_Crítico_Target, permitindo que os algoritmos aprendam padrões presentes nas variáveis preditoras — como segmento, região, agente causador, potencial do acidente e tempo de empresa — para posteriormente classificar novas ocorrências como "Sim" ou "Não".

O objetivo dos modelos será prever se determinado registro possui potencial para ser classificado como evento crítico.
